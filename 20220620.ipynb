## Saving Images

```python
im_rgb = im.convert("RGB")
```

```python
im_rgb.save("emma.jpg")
```

## 調換色層

```python
r, g, b = im_rgb.split()
im_rgb = Image.merge("RGB", (g, r, b))
```

```python
im_rgb
```

## Image Resize

```python
im. size
```

```python
im. size[0]
```

```python
im_rgb.resize((im. size[0]//4,im. size[1]//4))
```

```python
im_rgb = im_rgb.resize((im. size[0]//4,im. size[1]//4))
```

## Image Rotation
* transpose
* rotate

```python
im_rgb.transpose(Image.ROTATE_90)
```

```python
im_rgb.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
im_rgb.rotate(45)
```

```python
im_rgb.rotate(45, expand=True)
```

```python
im_rgb.rotate(60, expand =True, fillcolor = "yellow")
```

## Adding Text

```python
from PIL import ImageDraw
draw = ImageDraw.Draw(im_rgb)
```

```python
draw.text((0, 0),"Stay Back!!",(255,255,255))
```
